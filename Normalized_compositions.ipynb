```python
import json

# Load your JSON file
with open('admin.materials_cleaned_2.json', 'r') as file:
    data = json.load(file)

def normalize_composition(comp):
    total = sum(comp.values())
    if total == 0:
        return comp  # Edge case (empty composition)
    # Scale to atomic percentage (total = 100)
    print(comp.items())
    normalized = {element: (value / total) * 100 for element, value in comp.items()}
    # Round to 2 decimals
    normalized = {k: round(v, 2) for k, v in normalized.items()}
    return normalized

# Process each material
for entry in data:
    if 'composition' in entry:
        composition = entry['composition']
        total = sum(composition.values())
        # Performing the normalisation if total is significantly different from 100
        if abs(total - 100) > 1e-2:
            entry['composition'] = normalize_composition(composition)

# Save the updated JSON
output_file = 'admin.materials_cleaned_normalized.json'
with open(output_file, 'w') as file:
    json.dump(data, file, indent=2)

print(f"Normalized compositions saved to {output_file}")
```
